```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# --- 3 mm font size in points ---
mm_font = 2.5 / 25.4 * 72   # ≈ 8.5 pt

# --- Force matplotlib to use Arial ---
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.sans-serif'] = ['Arial']

# --- Set all font sizes ---
plt.rcParams['font.size'] = mm_font
plt.rcParams['axes.titlesize'] = mm_font
plt.rcParams['axes.labelsize'] = mm_font
plt.rcParams['xtick.labelsize'] = mm_font
plt.rcParams['ytick.labelsize'] = mm_font
plt.rcParams['legend.fontsize'] = mm_font
```

```python

H = np.loadtxt('data/color_analysis_H.csv',delimiter=',')
S = np.loadtxt('data/color_analysis_S.csv',delimiter=',')
V = np.loadtxt('data/color_analysis_V.csv',delimiter=',')

# 7. HISTOGRAMS (CLIPPED H, S, V)
fig, axs = plt.subplots(1, 3, figsize=(140/25.4, 50/25.4))  # ~210 mm × 70 mm

axs[0].hist(H, bins=60, color="gray", density = True)
axs[0].set_xlabel("Hue")
axs[0].set_ylabel("Probability density")
axs[0].set_xlim([0,180])
axs[0].set_xticks([0,60,120,180])
axs[0].set_ylim([0,0.25])
axs[0].set_yticks([0.,0.1,0.2,0.25])

axs[1].hist(S, bins=10, color="gray", density = True)
axs[1].set_xlabel("Saturation")
axs[1].set_xlim([0,1])
axs[1].set_xticks([0,0.25,0.5,0.75,1])
axs[1].set_ylim([0,45.])
axs[1].set_yticks([0.,15.,30.,45.])

#axs[0].set_xticks([0,.25,.5,180])

axs[2].hist(V, bins=20, color="gray", density = True)
axs[2].set_xlabel("Value")
axs[2].set_xlim([0,1])
axs[2].set_xticks([0,0.25,0.5,0.75,1])

axs[2].set_ylim([0,20.])
axs[2].set_yticks([0.,5.,10.,15.,20.])

plt.tight_layout(pad=0.1)
plt.savefig('PDFs/HSL.pdf', transparent=True)
plt.show()
```
