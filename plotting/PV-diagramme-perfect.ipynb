```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
import re
from colorPalette import colors
from scipy.interpolate import interp1d

mm_to_pt = 72 / 25.4
mm_to_inch = 1 / 25.4
font_size_mm = 3
font_size_pt = font_size_mm * mm_to_pt  # ≈ 8.5 pt
mpl.rcParams.update({
    "font.family": "sans-serif",
    "font.sans-serif": ["Arial"],
    "font.size": font_size_pt,
    "axes.labelsize": font_size_pt,
    "axes.titlesize": font_size_pt,
    "xtick.labelsize": font_size_pt,
    "ytick.labelsize": font_size_pt,
    "legend.fontsize": font_size_pt,
})
```

```python
n_bump='00'

csv_dir = '../data'
```

```python

PV_00_vol           = np.loadtxt(csv_dir+'/PV_00_vol.csv', delimiter =',')
PV_00_pressure_mean = np.loadtxt(csv_dir+'/PV_00_pressure_mean.csv', delimiter =',')
PV_00_pressure_std  = np.loadtxt(csv_dir+'/PV_00_pressure_std.csv', delimiter =',')

# Create figure with subplot
fig, ax = plt.subplots(figsize=(67.5 * mm_to_inch, 67.5 * mm_to_inch))

ax.plot(PV_00_vol, PV_00_pressure_mean, 'k-', label='Mean')
ax.fill_between(
    PV_00_vol,
    PV_00_pressure_mean - PV_00_pressure_std,
    PV_00_pressure_mean + PV_00_pressure_std,
    color='gray',
    alpha=0.3,
    label='Stddev'
)

ax.set_xlabel(r"Normalized volume, $\Delta V/V_\mathrm{init}$")
ax.set_ylabel(r"Normalized pressure, $\Delta P/P_\mathrm{cr}$ [-]")

ax.grid(True, alpha=0.3)
ax.legend()

plt.tight_layout()
plt.show()
```
