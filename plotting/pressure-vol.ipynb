```python
import numpy as np
import matplotlib.pyplot as plt

# unit conversions 
mm_to_inch = 1 / 25.4
font_mm = 2.5
font_pt = font_mm * 72 / 25.4   # mm -> points

# set font globally 
plt.rcParams["font.family"] = "Arial"

# load CSV 
data = np.loadtxt("../data/pressure-vol.csv", delimiter=",")

vol = data[:, 0]
pressure = data[:, 1]

# plot 
plt.figure(figsize=(67.5 * mm_to_inch, 67.5 * mm_to_inch))

plt.plot(vol/1000., pressure,'#215CAF')
plt.plot([0, 6.5/1000.], [102.4,102.4],'#B7352D')
         
plt.xlim([0,6.5/1000.])
plt.ylim([0,120])
plt.xlabel(r"Normalized volume, $\Delta V/V_\mathrm{init}$", fontsize=font_pt)
plt.ylabel(r"Normalized pressure, $\Delta P/P_\mathrm{cr}$ [-]", fontsize=font_pt)

plt.text(0.1/1000.,105,"Analytical buckling pressure",color='#B7352D', fontsize=font_pt)

plt.xticks(fontsize=font_pt)
plt.yticks(fontsize=font_pt)

plt.tight_layout(pad=0.2)
plt.show()
```
