```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os

MM_TO_PT = 72 / 25.4
font_size_mm = 2.5
font_size_pt = font_size_mm * MM_TO_PT  # ≈ 8.5 pt
mpl.rcParams.update({
    "font.family": "sans-serif",
    "font.sans-serif": ["Arial"],
    "font.size": font_size_pt,
    "axes.labelsize": font_size_pt,
    "axes.titlesize": font_size_pt,
    "xtick.labelsize": font_size_pt,
    "ytick.labelsize": font_size_pt,
    "legend.fontsize": font_size_pt,
})
```

```python
csv_dir = '../data'

# Load plot-ready data
df = pd.read_csv(csv_dir+'/thickness_azimuthal_profiles.csv')

# Plotting
fig_width = 88 / 25.4
fig_height = 35 / 25.4
xticks = np.linspace(0, 2 * np.pi, 9)

plt.figure(figsize=(fig_width, fig_height))
plt.plot(df['theta_rad'], df['t_over_R_vol_low'],
         label=r'$vol_\mathrm{low}$',  color='#627313', lw=1, alpha=0.75)
plt.plot(df['theta_rad'], df['t_over_R_vol_mid'],
         label=r'$vol_\mathrm{mid}$',  color='#C18A3A', lw=1, alpha=0.75)
plt.plot(df['theta_rad'], df['t_over_R_vol_high'],
         label=r'$vol_\mathrm{high}$', color='#B7352D', lw=1, alpha=0.75)

xticklabels = [r'$\theta=0$', r'$1/4$', r'$1/2$', r'$3/4$', r'$1$',
               r'$5/4$', r'$3/2$', r'$7/4$', r'$2$']

plt.xticks(ticks=xticks, labels=xticklabels)
plt.tick_params(axis='both', pad=2)

plt.xlim(0.0, np.pi * 2)
plt.ylim(0.0, 0.025)
plt.ylabel(r"Norm. thick., $t/R$", labelpad=1)
plt.xlabel(r"Angle $\theta$ ($\pi$)", labelpad=1)

plt.legend(
    ncol=3,
    loc="lower center",
    columnspacing=0.5,
    labelspacing=0.2,
    handletextpad=0.3,
    borderpad=0.1,
    borderaxespad=0.1,
    handlelength=1.0,
    frameon=False,
)
plt.tight_layout()
plt.show()
```
