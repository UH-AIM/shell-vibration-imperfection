```python
import cv2
import numpy as np
import matplotlib as mpl

import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
MM_TO_PT = 72 / 25.4
font_size_mm = 2.5
font_size_pt = font_size_mm * MM_TO_PT  # ≈ 8.5 pt

mpl.rcParams.update({
    "font.family": "sans-serif",
    "font.sans-serif": ["Arial"],
    "font.size": font_size_pt,
    "axes.labelsize": font_size_pt,
    "axes.titlesize": font_size_pt,
    "xtick.labelsize": font_size_pt,
    "ytick.labelsize": font_size_pt,
    "legend.fontsize": font_size_pt,
})
```

```python
import matplotlib.pyplot as plt

labels = ['t0-s1',
        't0-s2',
        't0-s3',
        't8-s1',
        't8-s2',
        't8-s3',
         ]
        
N = len(labels)

mask = np.triu(np.ones((N, N), dtype=bool))

corr = np.loadtxt('../data/exp_correlation.csv', delimiter=',')
corr = np.ma.array(corr, mask=~mask)
           
#  plot 
fig, ax = plt.subplots(figsize=(60/25.4, 80/25.4))

im = ax.imshow(
    corr,
    cmap="cividis",
    vmin=-1.0,
    vmax=1.0
)

ax.set_xticks(range(N))
ax.set_yticks(range(N))
ax.set_xticklabels(labels, rotation=90)
ax.set_yticklabels(labels)

# ---------- colorbar below plot ----------
divider = make_axes_locatable(ax)
cax = divider.append_axes("bottom", size="4%", pad=0.5)

cb = fig.colorbar(im, cax=cax, orientation="horizontal")
cb.set_label("Pearson correlation")

plt.tight_layout(pad=0.)
plt.show()
```
