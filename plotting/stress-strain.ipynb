```python
import numpy as np
import matplotlib.pyplot as plt

# --- unit conversions ---
mm_to_inch = 1 / 25.4
font_mm = 2.5
font_pt = font_mm * 72 / 25.4   # mm → points

# --- set font globally ---
plt.rcParams["font.family"] = "Arial"

# --- load CSV ---
data = np.loadtxt("../data/stress-strain.csv", delimiter=",")

strain = data[:, 0]
stress = data[:, 1]

# --- plot ---
plt.figure(figsize=(50 * mm_to_inch, 50 * mm_to_inch))

plt.plot(strain, stress, linewidth=2)
plt.xlim([0,25])
plt.xticks([0,5,10,15,20,25])
plt.ylim([0,30])
plt.yticks([0,5,10,15,20,25,30])
plt.xlabel("Strain", fontsize=font_pt)
plt.ylabel("Stress [KPa]", fontsize=font_pt)
plt.xticks(fontsize=font_pt)
plt.yticks(fontsize=font_pt)

plt.tight_layout(pad=0.2)
plt.show()
```
